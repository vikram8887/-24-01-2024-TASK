```python
import numpy as np, pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.gridspec as G
```

```python
 fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

G = gridspec.GridSpec(3, 3)
plt.subplot(G[0,:])
sns.violinplot(np.random.exponential(size=(50))).set_title('Violin Plot')

plt.subplot(3,3,4)
sns.histplot(np.random.exponential(size=(100))).set_title('histplot')

plt.subplot(3,3,5)
sns.kdeplot(np.random.exponential(size=(100))).set_title('Density Plot')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))), color='r', linestyle='--', label='Mean')
axes[1, 1].axvline(np.median(np.mean(np.random.exponential(size=(100)))), color='g',linestyle='--', label='Median')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))) + np.std(np.random.exponential(size=(100))), color='b', linestyle='--')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))) - np.std(np.random.exponential(size=(100))), color='b', linestyle='--')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))) + 2 * np.std(np.random.exponential(size=(100))), color='m', linestyle='--')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))) - 2 * np.std(np.random.exponential(size=(100))), color='m', linestyle='--')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))) + 3 * np.std(np.random.exponential(size=(100))), color='y', linestyle='--')
axes[1, 1].axvline(np.mean(np.random.exponential(size=(100))) - 3 * np.std(np.random.exponential(size=(100))), color='y', linestyle='--')


plt.subplot(3,3,6)
axes[1,2].scatter(np.random.exponential(size=(100)), np.random.exponential(size=(100)))
axes[1,2].set_title('scatter Plot')
 

plt.subplot(3,3,8)
data_countplot = np.random.choice(['Category A', 'Category B', 'Category C'], size=(100,))
sns.countplot(x=data_countplot, order=['Category A', 'Category B', 'Category C'], palette='viridis')
plt.title('count plot')

plt.subplot(3,3,9)
data_left = np.random.randint(1, 10, size=(5,))
data_right = np.random.randint(1, 10, size=(5,))
plt.title('butterfly plot')

plt.barh(np.arange(len(data_left)), data_left, color='blue', label='Left Wing')
plt.barh(np.arange(len(data_right)), -data_right, color='red', label='Right Wing')

plt.yticks(np.arange(len(data_left)), ['A', 'B', 'C', 'D', 'E'])
plt.xlabel('Values')
```
